#### 8. What are default arguments? Demonstrate with a function that calculates power.
* **`Default arguments:`**
* These are the values provided in a functional definition If the caller doesnt pass a value for that parameter the default is used
  

```python
# Power calculation using default argument
def power(base,exponent=2):
    """Returns base required to power of exponent
    (default is square)"""
    return base**exponent
```

```python
print(power(5))
print(power(3**2))
```

#### 10. Create a function that accepts arbitrary keyword arguments using **kwargs.

```python
def print_kwargs(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} = {value}")
```

```python
print_kwargs(name = 'Sandy',age=23,country = 'india')
```

#### 15. Write a function `analyze_numbers` that returns the sum and average of a list of numbers.

```python
def analyze_nums(lst):
    s = 0
    for i in lst:
        s += i
    avg = s / len(lst) if lst else 0  
    return s, avg

lst = [10,20,30,40,50,60,70,80]
print(analyze_nums(lst))
```

#### 19. Write a recursive function to calculate factorial of a number.

```python
def factorial(x):
    if x==0 or x==1:
        return 1
    else:
        return x*factorial(x-1)
print(factorial(10))
```

### 23. Write a function that modifies a global variable using `global` keyword.

```python
counter = 0

def increase_counter():
    global counter
    counter +=1
    print("counter inside function:", counter)

increase_counter()

print("counter outside function",counter)
```

#### 24. Write a function with a nested function and use `nonlocal` keyword to modify the outer variable.

```python
def outer_function():
    count = 0

    def inner_function():
        nonlocal count
        count+=1
        print("Inner count:",count)
    inner_function()
    print("Outer count:",count)
outer_function()
```

### 26. Write a docstring for a function `calculate_area(length, width)` that returns area of a rectangle.

```python
def cal_area(length, width):
    """Calculate the area of a rectangle.

    Parameters:
    length (float): The length of the rectangle.
    width (float): The width of the rectangle.

    Returns:
    float: The area of the rectangle (length x width).
    """
    return length * width
cal_area(10,20)
```

### 27. What are higher-order functions? Explain with an example using `map()` or `filter()`.

```python
def square(n):
    return n * n  # or n ** 2

numbers = [1, 2, 3, 4, 5]

squared_numbers = list(map(square, numbers))

print(squared_numbers)
```
